```python
#!/usr/bin/python
'''
Created on Sep 16, 2010
kNN: k Nearest Neighbors

Input:      inX: vector to compare to existing dataset (1xN)
            dataSet: size m data set of known vectors (NxM)
            labels: data set labels (1xM vector)
            k: number of neighbors to use for comparison (should be an odd number)
            
Output:     the most popular class label

@author: pbharrin
'''
#这样第输入方式，导致整个程序都不用numpy作为函数前缀（例如zeros）
from numpy import *
import operator #运算符模块，k近邻排序算法进行排序时用到这个模块提供的函数
from os import listdir #列出给定目录的文件名

#2-1 k近邻算法
#inX表示测试点；训练样本dataSet；labels训练样本的标签向量，即训练样本的名称；k近邻
def classify0(inX, dataSet, labels, k):
    dataSetSize = dataSet.shape[0]  #shape设置dataSet的属性（维度）shpe[0]取行数
    diffMat = tile(inX, (dataSetSize,1)) - dataSet #tile（numpy中函数）重复inx，行dataSetSize 列 ；计算一个测试点到所有训练样本的距离
    sqDiffMat = diffMat**2 #距离平方
    sqDistances = sqDiffMat.sum(axis=1) #对一个样本所有特征距离（列与列相加）求和（行代表样本，列代表每个样本的特征（因变量））
    distances = sqDistances**0.5 #后平方
    sortedDistIndicies = distances.argsort()     #对距离排序，并且返回距离的下标
    classCount={}          #构建字典
    for i in range(k):
        voteIlabel = labels[sortedDistIndicies[i]] #第k位是距离最近的样本点
        classCount[voteIlabel] = classCount.get(voteIlabel,0) + 1 #如果voteIlabel在classCount中就返回classCount[voteIlabel]；否则返回0
    #classCount.iteritems()按照第二个元素的次序对元祖进行排序，此处排序为逆排序
    #import operator  ； key=operator.itemgetter(1)得到字典中的第2项，并按照key排序；reverse翻转（降序排列）
    #classCount.iteritems()遍历字典；
    sortedClassCount = sorted(classCount.iteritems(), key=operator.itemgetter(1), reverse=True)
    return sortedClassCount[0][0] #得到第1个字典的第一个分量（即key值）

def createDataSet():
    group = array([[1.0,1.1],[1.0,1.0],[0,0],[0,0.1]]) #创建数组
    labels = ['A','A','B','B'] #创建标签
    return group, labels
'''
1 将文本记录到numpy的解析程序;
2 classLabelVector.append(int(listFromLine[-1]))
里面的int去掉，明显不能这样转换，因为都是文字 ;
3 zeros是numpy中的函数，由于采用from numpy import *输入方式，
所以zeros不用前缀numpy
 
'''      
def file2matrix(filename):
    fr = open(filename) #打开文件
    numberOfLines = len(fr.readlines())         #get the number of lines in the file
    returnMat = zeros((numberOfLines,3))        #以0填充的特征矩阵
    classLabelVector = []                       #存储矩阵的列名称  
    fr = open(filename)
    index = 0
    for line in fr.readlines():
        line = line.strip() #截取所有回车字符
        listFromLine = line.split('\t')    #用tab(\t)键得到line的整行数据分割成一个元素列表。
        returnMat[index,:] = listFromLine[0:3] #每行取前3个元素
        classLabelVector.append(listFromLine[-1])#将最后一列添加到向量中
        index += 1
    return returnMat,classLabelVector

'''
归一化特征值，使数据在0-1之间
在Numpy库中linalg.solve(matA, matB)矩阵除法
normDataSet为每一个点离最小值点的距离, 
#ranges最值点最小值点的距离, 
#minVals 最小点

'''    
def autoNorm(dataSet):
    minVals = dataSet.min(0)    #取每列的最小值第0位
    maxVals = dataSet.max(0)    #取每列的最大值第0位
    ranges = maxVals - minVals
    normDataSet = zeros(shape(dataSet)) #shape取得数据的维度dim，并设置矩阵为0
    m = dataSet.shape[0] #shape[0]取行数，shape[1]取列数
    normDataSet = dataSet - tile(minVals, (m,1))
    normDataSet = normDataSet/tile(ranges, (m,1))   #element wise divide
    return normDataSet, ranges, minVals


'''
分类器针对约会网站的测试代码
'''   
def datingClassTest():
    hoRatio = 0.1      #hold out 10%
    datingDataMat,datingLabels = file2matrix('datingTestSet2.txt')       #load data setfrom file
    normMat, ranges, minVals = autoNorm(datingDataMat) 
    m = normMat.shape[0] 
    numTestVecs = int(m*hoRatio) #取前10%行作为测试集
    errorCount = 0.0
    for i in range(numTestVecs):
        classifierResult = classify0(normMat[i,:],normMat[numTestVecs:m,:],datingLabels[numTestVecs:m],3) #返回的是字符，所以不能用%d
        print "the classifier came back with: %s, the real answer is: %s" % (classifierResult, datingLabels[i])
        if (classifierResult != datingLabels[i]): errorCount += 1.0
    print "the total error rate is: %f" % (errorCount/float(numTestVecs))
    print errorCount

#在python中不能用转化float，在ipython中可以
def classifyPerson():
    resultList = ['not at all', 'in small doses','in large doses']
    #\为换行符
    percentTats = float(raw_input(\
    "percentage of time spent playing video games?"))
    ffMile = float(raw_input("frequent fliter miles earned per year?"))
    iceCream = float(raw_input("liters of ice cream consumed per year?"))
    datingDataMat, datingLables = file2matrix('datingTestSet2.txt')
    normMat, ranges, minVals = autoNorm(datingDataMat)
    inArr = array([ffMile, percentTats, iceCream])
    classifierResult = classify0((inArr-minVals)/ranges, normMat, datingLables,3)
    print "you will probably like this person:", \
    resultList[int(classifierResult) - 1]

#2.3.1准备数据：将图像转化为测试向量    
def img2vector(filename):
    returnVect = zeros((1,1024)) #构造行1列1024的0数组
    fr = open(filename)
    for i in range(32):
        lineStr = fr.readline()  #取前32个字符
        for j in range(32):
            returnVect[0,32*i+j] = int(lineStr[j]) #
    return returnVect

#测试算法：使用k近邻识别手写数字
def handwritingClassTest():
    hwLabels = [] #构建一个列表
    trainingFileList = listdir('trainingDigits')           #得到ch02路劲下trainingDigits文件夹下的文件名
    m = len(trainingFileList) #文件名的长度
    trainingMat = zeros((m,1024)) 
    for i in range(m):
        fileNameStr = trainingFileList[i] #每一个元素都放在fileNameStr中
        fileStr = fileNameStr.split('.')[0]     #0_0.txt以.为分隔符，取第一部分得到.txt前面的名称
        classNumStr = int(fileStr.split('_')[0]) #解析出分类数字
        hwLabels.append(classNumStr) #把名字加在后面
        trainingMat[i,:] = img2vector('trainingDigits/%s' % fileNameStr) #每行储存一个图像
    testFileList = listdir('testDigits')        #iterate through the test set
    errorCount = 0.0
    mTest = len(testFileList)
    for i in range(mTest):
        fileNameStr = testFileList[i]
        fileStr = fileNameStr.split('.')[0]     #take off .txt
        classNumStr = int(fileStr.split('_')[0])
        vectorUnderTest = img2vector('testDigits/%s' % fileNameStr)
        classifierResult = classify0(vectorUnderTest, trainingMat, hwLabels, 3)
        print "the classifier came back with: %d, the real answer is: %d" % (classifierResult, classNumStr)
        if (classifierResult != classNumStr): errorCount += 1.0
    print "\nthe total number of errors is: %d" % errorCount #\n换行，
    print "\nthe total error rate is: %f" % (errorCount/float(mTest))
```
